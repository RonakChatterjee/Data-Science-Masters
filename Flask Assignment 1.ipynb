```python
#Ans 3:
"""
In Flask, app routing refers to the process of defining the URL endpoints (routes) and associating them with corresponding view functions that handle 
the incoming requests. It allows you to map specific URLs to specific functions in your Flask application.

The routing mechanism in Flask is achieved using the @app.route() decorator, which is applied to a Python function to bind it to a specific URL. 
The decorator takes the URL pattern as an argument and associates it with the decorated function. When a user visits a particular URL, Flask matches 
it with the defined routes and invokes the associated function to generate the appropriate response.
"""
```
