```python
"""
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
"""
```

```python
#Ans 4:
"""
The LookupError exception in Python is the base class for all exceptions that are raised when an index or a key is not found for a 
sequence or dictionary respectively.
"""
```

```python
#KeyError
mydict = {
    'name' : 'Ronak',
    'age' : 23,
    'course' : 'Data Science Masters'
}

try:
    mydict['rollno']
except KeyError as e:
    print("We encountered KeyError")

# print(mydict['rollno'])
```

```python
#IndexError
mylist = [1,2,3]

try:
    mylist[3]
except IndexError as e:
    print("We encountered IndexError")
```
