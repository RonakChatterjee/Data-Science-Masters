```python
#Ans 5:
"""
Primary Key: It is the attrbute/column of the table that is always unique and not null. It helps us to get a row from the table uniquely.

Foreign Key: A foreign key is generally used to build a relationship between the two tables. 

Let's see the below example:

Table1: With student details

Create Table Student(
        RollID int primary key,
        name varchar(20),
        address varchar(50)
        );
      
Table2: With Course details

Create Table Courses(
        CourseID int primary key,
        Course Name varchar(30),
        RollID int references Student(RollID)
        ):

Here RollID in Table1 behaves as the primary key. And in Table2 RollID acts as a foreign key and helps to maintain a relation between Student and 
Courses table.

"""
```
