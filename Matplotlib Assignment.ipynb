```python
'''Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.
'''
```

```python
#Ans2:
'''A scatter plot is a visual representation of how two variables relate to each other and is represented as dots in a graph.
Below is an example
'''
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot')
```

```python
'''Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])'''
```

```python
#Ans3:
'''The subplot() fucnction provides a way to plot multiple plots on a single figure based on the number of rows and columns provided.
Below is the example:
'''
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.subplot(2,2,1)
plt.plot(np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500]), c = 'r')
plt.title('Graph1')

plt.subplot(2,2,2)
plt.plot(np.array([0, 1, 2, 3, 4, 5]) , np.array([50, 20, 40, 20, 60, 70]), c = 'g')
plt.title('Graph2')

plt.subplot(2,2,3)
plt.plot(np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60]), c = 'b')
plt.title('Graph3')

plt.subplot(2,2,4)
plt.plot(np.array([0, 1, 2, 3, 4, 5]) ,np.array([200, 350, 250, 550, 450, 150]), c = 'y')
plt.title('Graph4')
```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])'''
```

```python
#Ans4:
'''A barplot is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. Each entity 
of the categoric variable is represented as a bar. The size of the bar represents its numeric value.'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit,color='g')
```

```python
plt.barh(company,profit)
```

```python
'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)'''
```

```python
#Ans5:
'''Box plots provide a quick visual summary of the variability of values in a dataset. They show the median, upper and lower quartiles,
 minimum and maximum values, and any outliers in the dataset. Outliers can reveal mistakes or unusual occurrences in data.'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1,box2]

plt.boxplot(data,vert=True,patch_artist=True)
```
